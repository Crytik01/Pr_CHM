```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt

xi = np.array ([0.15,0.20,0.25,0.30,0.35,0.40,0.45,0.50,0.55,0.60,0.65])
yi = np.array([0.8607,0.8187,0.7788,0.7408,0.7046,0.6703,0.6376,0.6065,0.5769,0.5488,0.5220])

x =np.array([0.151,0.223,0.252,0.505,0.554,0.457])

def interpolation(x, y, x0):
    
    n = len(x)
    f = np.zeros((n, n))
    f[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            f[i, j] = (f[i+1, j-1] - f[i, j-1]) / (x[i+j] - x[i])
    ans = 0
    for j in range(n):
        prod = f[0, j]
        for i in range(j):
            prod *= (x0 - x[i])
        ans += prod
        
    return ans

y = interpolation(xi, yi, x)

print(f"f({x}) = {y}")

xx = np.linspace(np.min(x), np.max(x), 100)
yy = np.zeros_like(xx)

for i in range(len(xx)):
    yy[i] = interpolation(x, y, xx[i])
    
plt.plot(x, y, 'o', label='Дані точки')
plt.plot(xx, yy, label='багаточлен Ньютона')
plt.title("Графік інтерполяційної функції Ньютона")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
